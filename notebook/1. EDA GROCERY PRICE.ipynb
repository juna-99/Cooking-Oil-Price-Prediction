## PREMISE

```python
URL_DATA = 'https://storage.data.gov.my/pricecatcher/lookup_premise.parquet'

premise = pd.read_parquet(URL_DATA)
premise
```

```python
premise.isnull().sum()
```

```python
# Replace 'None' with NaN
premise.replace(['None','',',',',,'],np.nan, inplace=True)

# Drop rows where all values are 'None'
premise = premise.dropna(how='all', subset=premise.columns[1:])

# Display the cleaned DataFrame
premise
```

```python
premise.isnull().sum()
```

```python
premise.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the frequency of each premise type
premise_type_counts = premise['premise_type'].value_counts()

# Plotting the distribution
plt.figure(figsize=(12, 6))
sns.barplot(x=premise_type_counts.values, y=premise_type_counts.index, palette="viridis")
plt.title('Distribution of Premise Types')
plt.xlabel('Count')
plt.ylabel('Premise Type')
plt.grid(axis='x')
plt.show()

premise_type_counts
```

```python
# Count of different premise types in each state
premise_type_by_state = premise.groupby(['state', 'premise_type']).size().unstack().fillna(0)

# Plotting the count of different premise types in each state
premise_type_by_state.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='viridis')
plt.title('Count of Different Premise Types in Each State')
plt.xlabel('State')
plt.ylabel('Count of Premises')
plt.xticks(rotation=45)
plt.legend(title='Premise Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
columns_to_drop1 = ['premise', 'address','district']
premise.drop(columns=columns_to_drop1, inplace=True)
premise
```

```python
premise.dtypes
```

```python
premise['premise_code'] = premise['premise_code'].astype(int)
```

```python
premise
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grouping the data by state and counting the number of premises in each state
state_counts = premise.groupby('state')['premise_code'].count().reset_index()
state_counts.columns = ['State', 'Premise Count']

# Sorting the states by premise count (optional)
state_counts = state_counts.sort_values('Premise Count', ascending=False)

# Plotting the bar chart using Seaborn
plt.figure(figsize=(10, 7))
sns.barplot(x='State', y='Premise Count', data=state_counts, palette='viridis')
plt.xlabel('State')
plt.ylabel('Number of Premises')
plt.title('Store Distribution by State in Malaysia')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
for state in states:
    state_data = data[data['state'] == state]
    
    # Aggregate the data: average price per day
    aggregated_data = state_data.groupby('date')['price'].mean().reset_index()
    
    # Plot the data
    plt.figure(figsize=(10, 5))
    plt.plot(aggregated_data['date'], aggregated_data['price'])
    plt.title(f'Price Data for {state}')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.grid(True)
    plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder

# One-hot encode the premise_type column
encoder = OneHotEncoder(drop='first')
premise_type_encoded = encoder.fit_transform(data[['premise_type']]).toarray()

# Create a DataFrame with the encoded premise types
premise_type_encoded_df = pd.DataFrame(premise_type_encoded, columns=encoder.get_feature_names_out(['premise_type']))

# Combine the original data with the encoded columns
data_encoded = pd.concat([data[['price']], premise_type_encoded_df], axis=1)

# Calculate the correlation matrix
correlation_matrix = data_encoded.corr()

import ace_tools as tools; tools.display_dataframe_to_user(name="Correlation Matrix", dataframe=correlation_matrix)

correlation_matrix
```
